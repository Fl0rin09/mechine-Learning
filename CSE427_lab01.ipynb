# **NumPy**

```python
import numpy as np

# 1
a = np.random.randint(1, 101, 20)
b = np.random.randint(1, 101, 20)
print("Array a:", a)
print("Array b:", b)
```

```python
# 2
a = a.reshape(5, 4)
b = b.reshape(5, 4)
print("\nMatrix a:\n", a)
print("\nMatrix b:\n", b)
```

```python
# 3
a_scaled = 2 * a
b_scaled = 3 * b
print("\nScalar Multiplication of a (x2):\n", a_scaled)
print("\nScalar Multiplication of b (x3):\n", b_scaled)
```

```python
# 4
b[0][0] = 1
c = a @ b.T
print("\nMatrix c (a × bᵀ):\n", c)
```

```python
# 5
print("\nColumn 2 and 3 of matrix c:\n", c[:, 1:3])
print("\nFull matrix c:\n", c[:])
```

```python

# 6
max_val = c.max()
max_index = np.unravel_index(c.argmax(), c.shape)
print("\nLargest value in c:", max_val)
print("Index of largest value:", max_index)
```

```python
# 7
c_flat = c.flatten()
print("\nFlattened c (1D):\n", c_flat)
```

```python
import pandas as pd
url = "/content/subject_scores.csv"
df = pd.read_csv(url)

print("\nFirst few rows:\n", df.head())
print("\nDataset Info:")
df.info()
print("\nNull values:\n", df.isnull().sum())
df = df.fillna(df.mean(numeric_only=True))
print("\nNull values after fill:\n", df.isnull().sum())
if "Math" in df.columns:
    print("\nMath Scores:\n", df["Math"])
else:
    print("\n(No column named 'Math' found. Check dataset columns.)")
```

# ***Matplotlib***

```python
#Task-1
import matplotlib.pyplot as plt
temperature_dhaka = [25,34,21,45,28,6,43,18,7,2]
humidity_dhaka = [28,25,29,20,26,50,19,29,52,55]

plt.figure(figsize=(6,4))
plt.scatter(temperature_dhaka, humidity_dhaka, marker='*', color='red')
plt.title("Temperature vs Humidity")
plt.xlabel("Temperature (°C)")
plt.ylabel("Humidity (%)")
plt.show()
```

```python
#Task-5
import matplotlib.pyplot as plt

temperature_dhaka = [25,34,21,45,28,6,43,18,7,2]
humidity_dhaka = [28,25,29,20,26,50,19,29,52,55]
study_hours = [2,3,4,4,5,6,7,7,8,9,9,10,11,11,12]
marks_study = [6,10,15,20,34,44,55,60,55,67,70,80,90,99,100]
subjects = ['Maths', 'English', 'Science', 'Physics', 'Computer']
marks_subjects = [89, 90, 45, 78, 99]
colors = ['blue', 'red', 'green', 'purple', 'yellow']

plt.figure(figsize=(12,10))

plt.subplot(2,2,1)
plt.scatter(temperature_dhaka, humidity_dhaka, marker='*', color='red')
plt.title("Temp vs Humidity")

plt.subplot(2,2,2)
plt.plot(study_hours, marks_study)
plt.title("Study Hours vs Marks")

plt.subplot(2,2,3)
plt.barh(subjects, marks_subjects)
plt.title("Subject Scores ")

plt.subplot(2,2,4)
plt.bar(subjects, marks_subjects, color=colors)
plt.title("Subject Scores ")

plt.tight_layout()
plt.show()
```
